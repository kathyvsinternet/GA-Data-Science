```python
# Set up Python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load casting data

df_casting = pd.read_csv('../project_movies_data/movies_casting_raw.csv', sep = ';')
```

```python
df_casting.head()
```

```python
# Correct column for screenwriter and drop wrong column

df_casting['screenwriter_name'] = df_casting['screeplay_name']
df_casting.drop(['screeplay_name'], axis = 1, inplace = True)
```

```python
# Convert data types

df_casting[['actor2_gender', 'director_gender']] = df_casting[['actor2_gender', 'director_gender']].fillna(0).apply(lambda x: x.astype('int64'))
```

```python
# Lowercase names

df_casting[['actor1_name', 'actor2_name', 'actor3_name', 'actor4_name', 'actor5_name', 'director_name', 'producer_name', 'screenwriter_name', 'editor_name']] = df_casting[['actor1_name', 'actor2_name', 'actor3_name', 'actor4_name', 'actor5_name', 'director_name', 'producer_name', 'screenwriter_name', 'editor_name']].apply(lambda x: x.str.lower())
```
